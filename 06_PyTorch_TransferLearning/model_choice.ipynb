```python
import pandas as pd
```

```python
df = pd.read_csv('./ModelChoice.csv', sep=';')
```

```python
list_columns = ['Acc@1', 'Acc@5', 'Params']
for col in list_columns:
    df[col] = df[col].str.replace(',', '.', regex=True)

df[list_columns] = df[list_columns].astype('float32')
df
```

```python
df.info()
```

```python
best_models = df.loc[df['Acc@1'] > df['Acc@1'].quantile(0.75), :].copy()
best_models['Efficiency'] = best_models['Acc@1'] / best_models['Params']
best_models
```
